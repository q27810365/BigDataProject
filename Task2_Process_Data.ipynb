#### Feature Columns
* Date – The date when the sample is collected.
* Number of rooms – number of rooms in the house.
* Security level of the community – the higher the safer.
* Residence space – square feet area of the living rooms.
* Building space – square feet area of the whole building.
* Noise level – the lower the value, the greater the noise.
* Waterfront – If the house has water front or not.
* View – Number of viewings before the house is sold.
* Air quality level – the higher the value, the better the air quality.
* Aboveground space – square feet area of the above house.
* Basement space – square feet area of the basement in the house.
* Building year – the year in which the house was built.
* Decoration year – the year in which the house was decorated.
* District – the address of the house.
* City – the city in which the house is located.
* Zip code – the zip code of the house.
* Region – the region of the house.
* Exchange rate – when the house is sold, the exchange rate between the US dollar and the Hong Kong dollar.
* Unit price of residence space – the unit price of residence space (US dollars).
* Unit price of building space – the unit price of building space (US dollars).
* Total cost – the total price of residence and building space (Hong Kong dollars).

## Data Processing
* Convert total costs to classes
* Simplify Date (Year and Time are nonsense)
* Convert District to latitude and longitude
* Remove Region (All USA)
* Remove unit price of residence space and unit price of building space
* Convert dtype:Object (String) to number

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# read Train_Data.csv
df = pd.read_csv('data/Train_Data.csv')
df.head(5)
```

```python
# conver total costs to classes
for index,row in df.iterrows():
    total_cost = row['total cost']
    if total_cost >= 700000:
        row['total cost'] = 4
    elif 700000 > total_cost >= 500000:
        row['total cost'] = 3
    elif 500000 > total_cost >= 300000:
        row['total cost'] = 2
    elif 300000 > total_cost >= 0:
        row['total cost'] = 1
    df.iloc[index] = row
# convert from float to int
df['total cost'] = df['total cost'].astype(int)
df.head(5)
```

```python
# Simplify Date, create 2 new rows Month and Day, remove Date
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].apply(lambda date:date.month)
df['day'] = df['date'].apply(lambda date:date.day)
df = df.drop('date',axis=1)
df.head(5)
```

```python
# Convert District to latitude and longitude and remove it, use library [GeoPy]
from geopy.geocoders import Bing
df['latitude'] = ''
df['longitude'] = ''
# Transactions limited! CAREFULLY RUN THE CODE IN THE NEXT CELL
geolocator = Bing('AtlROln1P4u7I9A7cRfUhfmcTA_lZ7n24mOxkDqRoTFkriiG4M1F1I36nTPbQnUG')
# test
location = geolocator.geocode('5944-6082 Delridge Way SW')

print((float(location.latitude), float(location.longitude)))
```

```python
# Convert dtype:Object (String) to number
# One-Hot Encode
from sklearn import preprocessing
encoder = preprocessing.OneHotEncoder(handle_unknown='ignore')
encoder.fit(df[["city","zip code"]])
# Get the feature name after encoding
feature_names=encoder.get_feature_names_out(df[["city","zip code"]].columns)
df_encoded=pd.DataFrame(encoder.transform(df[["city","zip code"]]).toarray(),columns=feature_names)

df = pd.concat([df,df_encoded],axis=1)
df.drop(df[["city","zip code"]],axis = 1 , inplace=True)

df.head(5)
```
